### **1. Read Mode (`"r"`)**

* Opens a file **only for reading**.
* The file **must exist**, otherwise Python will give an error.
* Example:

```python
file = open("example.txt", "r")  # open file for reading
content = file.read()            # read all content
print(content)
file.close()
```

* Variants:

  * `"r"` → read text (default)
  * `"rb"` → read binary file (like images)

---

### **2. Write Mode (`"w"`)**

* Opens a file **for writing**.
* If the file **does not exist**, it will **create a new file**.
* If the file **already exists**, it will **erase all existing content**.
* Example:

```python
file = open("example.txt", "w")  # open file for writing
file.write("Hello World!\n")     # write text to file
file.write("Python is fun.")
file.close()
```

* Variants:

  * `"w"` → write text
  * `"wb"` → write binary file

---

### **3. Append Mode (`"a"`)**

* Opens a file **to add new data at the end** without deleting existing content.
* Example:

```python
file = open("example.txt", "a")  
file.write("\nAdding a new line.")  # adds to end
file.close()
```

---

### **4. Read & Write Mode (`"r+"`)**

* Opens a file for **both reading and writing**.
* File **must exist**.

```python
file = open("example.txt", "r+")
print(file.read())        # read content
file.write("\nNew text")  # write at current cursor
file.close()
```

---

✅ **Quick Summary Table:**

| Mode   | Action       | File must exist? | Overwrites content? |
| ------ | ------------ | ---------------- | ------------------- |
| `"r"`  | Read         | Yes              | No                  |
| `"w"`  | Write        | No               | Yes                 |
| `"a"`  | Append       | No               | No                  |
| `"r+"` | Read & Write | Yes              | No                  |

### **`w+` Mode**

* Opens a file **for both writing and reading**.
* If the file **does not exist**, Python will **create a new file**.
* If the file **already exists**, **all its content will be erased** when opened.
* You can **write to the file** and **read from it**, but be careful with the cursor position.

---

### **Example:**

```python
# Open file in w+ mode
file = open("example.txt", "w+")

# Write to the file
file.write("Hello World!\n")
file.write("Python is fun.\n")

# Move cursor to the beginning to read from start
file.seek(0)

# Read content
content = file.read()
print(content)

file.close()
```

**Output:**

```
Hello World!
Python is fun.
```

---

### **Key Points about `w+`**

1. **Erases existing content** when file opens.
2. Can **read and write** in the same file.
3. To read after writing, you often need **`file.seek(0)`** to move the cursor to the start.
4. Creates the file if it doesn’t exist.
