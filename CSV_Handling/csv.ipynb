```python
import csv
csv_file_path = "./sample_persion_data/person_data.csv"

data = [
    {
        "name": "Kamal",
        "age": 32,
        "city": "Colombo"
    },
    {
        "name": "Anjali",
        "age": 25,
        "city": "Kandy"
    },
    {
        "name": "Suresh",
        "age": 45,
        "city": "Galle"
    }
]

filenames = ["name", "age", "city"]

with open(csv_file_path, "w", newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=filenames)
    csv_writer.writeheader()
    csv_writer.writerows(data)
```
