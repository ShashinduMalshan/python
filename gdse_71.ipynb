

# Identity Operators in Python

Identity operators are used to compare the memory locations of two objects.  
They check whether two variables refer to the same object in memory, not just if their values are equal.

## Identity Operators

### `is`

Returns `True` if two variables point to the same object.

```python
x = [1, 2, 3]
y = x
print(x is y)   # True (both refer to same object)
````

### `is not`

Returns `True` if two variables do not point to the same object.

```python
x = [1, 2, 3]
y = [1, 2, 3]
print(x is not y)  # True (different objects, even though values are same)
```

## Difference Between `==` and `is`

* `==` → compares values.
* `is` → compares object identity (memory address).

```python
a = [10, 20]
b = [10, 20]
print(a == b)   # True → values are equal
print(a is b)   # False → different objects in memory
```

## Special Case: Integer Caching (Pooling)

Python caches integers from **-5 to 256** for performance.
These integers are stored in a pool, so variables with the same value in this range will refer to the same memory location.

This is why identity checks with `is` may return `True` within this range.

```python
x = 15
y = 15
print(x is y)   # True → same cached object

x = 256
y = 256
print(x is y)   # True → still cached

x = 257
y = 257
print(x is y)   # False → different objects
```

This optimization is often called **integer pooling** or **object caching**.

# Bitwise Operators in Python

## `&` → AND
Performs **bitwise AND** on two numbers. Each bit is 1 if **both corresponding bits are 1**.

```python
a = 5      # 0b0101
b = 3      # 0b0011
print(a & b)  # 1 → 0b0001
```

---

## `|` → OR
Performs **bitwise OR** on two numbers. Each bit is 1 if **at least one corresponding bit is 1**.

```python
a = 5      # 0b0101
b = 3      # 0b0011
print(a | b)  # 7 → 0b0111
```

---

## `^` → XOR
Performs **bitwise XOR** (exclusive OR). Each bit is 1 if **only one of the bits is 1**.

```python
a = 5      # 0b0101
b = 3      # 0b0011
print(a ^ b)  # 6 → 0b0110
```

---

## `~` → NOT
Performs **bitwise NOT** (one's complement). Flips all bits of the number.

```python
a = 5      # 0b0101
print(~a)   # -6 → -(5+1)
```

---

## `<<` → Left Shift
Shifts bits to the left by a specified number of positions. Fills right side with 0.

```python
a = 5      # 0b0101
print(a << 1)  # 10 → 0b1010
```

---

## `>>` → Right Shift
Shifts bits to the right by a specified number of positions. Fills left side with 0 (or 1 for negative numbers in Python).

```python
a = 5      # 0b0101
print(a >> 1)  # 2 → 0b0010
```

### Python `match-case` Example: Grade Evaluation

This Python code demonstrates how to use `match-case` (Python's version of `switch-case`) to evaluate a student's grade.

```python
# Input grade from user
grade = input("Enter your grade (A, B, C, D, F): ").upper()

# Match-case statement
match grade:
    case "A":
        print("Excellent!")
    case "B":
        print("Good job!")
    case "C":
        print("Average performance")
    case "D":
        print("Needs improvement")
    case "F":
        print("Fail")
    case _:
        print("Invalid grade entered")  # Default case
````

## Explanation

* **`match grade:`**
  This starts the `match-case` block, checking the value of the variable `grade`.

* **`case "A":`, `case "B":`, etc.**
  These are individual cases. If `grade` matches the value in a `case`, the corresponding block of code runs.

* **`case _:`**
  This is the **default case**. The underscore `_` acts like a wildcard, meaning it will match **any value that doesn’t match the previous cases**. It's similar to the `default` case in other languages like Java or C++.

* **`.upper()`**
  Converts the input to uppercase so that both `a` and `A` are treated the same.

### Step 1: `match number`

* `match` is checking the **structure and content of the list** `number`.
* Python’s `match-case` can **pattern match** lists, tuples, or other sequences.

---

### Step 2: `case [x, y]`

* This pattern matches a list **with exactly 2 elements**.
* `x` gets the first element, `y` gets the second.
* If `number` had 2 elements, it would print `x * y`.

Example:

```python
number = [4, 3]
# matches [x, y] -> prints 4 * 3 = 12
```

---

### Step 3: `case [x, y, z]`

* This pattern matches a list **with exactly 3 elements**.
* `x, y, z` get assigned to the first, second, and third elements of the list.
* For your list `[4, 3, 5]`, this matches and prints:

```python
4 + 3 + 5 = 12
```

---

### Step 4: Order matters

* Python checks `case` statements **top to bottom**.
* If `[x, y]` were **first**, and the list had 3 elements, it **would not match** `[x, y]` because it requires exactly 2 elements.
* Then Python moves to `[x, y, z]` and matches correctly.

---

### ✅ Summary:

* `match-case` can **pattern match lists** by length and position of elements.
* `case [x, y]` → only matches **2-element lists**.
* `case [x, y, z]` → only matches **3-element lists**.
* The variables `x, y, z` are automatically **assigned the values** from the list.

## 🔁 What is a `for` loop in Python?

A **`for` loop** is used to repeat a block of code a specific number of times or to iterate over a sequence like a list, string, tuple, or range.

---

## ✅ Syntax of a `for` loop

```python
for variable in sequence:
    # code block
```

* `variable` → represents each item from the sequence.
* `sequence` → something that you can loop through (like `range`, list, or string).
* The code block inside the loop runs once for each item.

---

## ✅ Example

```python
for i in range(5):
    print(i)
```

### Output:

```
0
1
2
3
4
```

Here, `range(5)` generates numbers from `0` to `4`.

---

## ✅ Rules of `for` loop in Python

| Rule                           | Description                                                       |
| ------------------------------ | ----------------------------------------------------------------- |
| Indentation                    | The code inside the loop must be indented.                        |
| Sequence required              | Must loop over a sequence: `range`, list, tuple, string, etc.     |
| Stops automatically            | Loop ends after finishing the last item in the sequence.          |
| Variable updates automatically | Loop variable changes automatically — no need to update manually. |
| Can use `break`                | Stops loop early.                                                 |
| Can use `continue`             | Skips current iteration.                                          |

## 🔍 Explanation

### ✅ First Loop – **Efficient Answer**

```python

    print(i)
```

* This loop prints numbers divisible by **7** from **7 to 99**.
* It uses `range(7, 100, 7)`:

  * `7` → start
  * `100` → end (not included)
  * `7` → step (count jumps by 7)
* So it prints only required numbers → **faster and more efficient** ✅

---

### 🐢 Second Loop – **Normal Answer**

```python
for i in range(7, 100):
    if i % 7 == 0:
        print(i)
```

* This also prints numbers divisible by 7 from 7 to 99.
* But it checks **every number one by one** with condition `i % 7 == 0`
* Slower because it makes **93 checks** (from 7 to 99) ❌
* Works correctly but **not efficient**.

---

## ⚙️ Why the First Loop is Efficient?

| Feature         | Efficient Loop ✅ | Normal Loop ❌                  |
| --------------- | ---------------- | ------------------------------ |
| Checks per loop | 0 conditions     | Checks `i % 7 == 0` every time |
| Loop steps      | Jumps by 7       | Goes 1 by 1                    |
| Speed           | Faster           | Slower                         |
| CPU effort      | Low              | Medium                         |

---

## ✅ Summary

* Use **step** in `range()` when possible → cleaner + faster.
* Use `if` inside loop only when necessary.

Protected members keep internal logic separate from public interfaces.

# 🔍 Name Mangling (Very Important)

Python secretly renames:

```python
self.__name
```

to:

```python
self._Dog__name
```

This prevents accidental access.

So the variable exists, but with a different name.

You must use a **method** inside the class:

```python
class Dog:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

dog_1 = Dog("Tommy")
print(dog_1.get_name())   # Tommy
```

# ⚠️ Can you still access it from outside? (Yes, but not recommended)

You can access using the mangled name:

```python
print(dog_1._Dog__name)   # Tommy
```

But this is **not recommended** and breaks OOP rules.

# **What is JSON?**

* **JSON** stands for **JavaScript Object Notation**.
* It is a **lightweight data format** used for **storing and exchanging data**.
* JSON is **text-based** and **easy for humans to read and write**.
* It is widely used in **web applications** to send data between a server and a client.

---

### **JSON Structure**

JSON is made of **key-value pairs**, similar to Python dictionaries.

#### Example:

```json
{
  "name": "Shasidu",
  "age": 17,
  "is_student": true,
  "courses": ["Python", "Java", "Math"],
  "address": {
    "city": "Colombo",
    "country": "Sri Lanka"
  }
}
```

* `"name"`, `"age"`, `"is_student"` → **keys**
* `"Shasidu"`, `17`, `true` → **values**
* Values can be:

  * String → `"Hello"`
  * Number → `123`
  * Boolean → `true` / `false`
  * Array → `["Python", "Java"]`
  * Object → `{ "key": "value" }`
  * `null` → represents no value

---

### **JSON in Python**

* In Python, JSON is handled using the **`json` module**.
* **Convert Python dictionary to JSON** (serialization):

```python
import json

data = {"name": "Shasidu", "age": 17}
json_string = json.dumps(data)  # Convert dict to JSON string
print(json_string)
```

* **Convert JSON to Python dictionary** (deserialization):

```python
import json

json_string = '{"name": "Shasidu", "age": 17}'
data = json.loads(json_string)  # Convert JSON string to dict
print(data["name"])
```

---

✅ **Summary**:

* JSON = text format for exchanging data.
* Easy to read, write, and parse.
* Works with almost all programming languages, not just JavaScript.

---



---

### **1. Data Type**

* **Python dictionary (`dict`)**: A **Python object** used to store key-value pairs.
* **JSON**: A **text format** (string) used to represent data. It is **language-independent**.

```python

```

Example in JSON (as string):

```json

  "name": "Shasidu",

### **2. Syntax Differences**

| Feature        | Python Dict                             | JSON                         |
| -------------- | --------------------------------------- | ---------------------------- |
| Keys           | Can be strings, numbers, or tuples      | Must be strings (in quotes)  |
| Strings        | Can use single `'` or double `"` quotes | Must use double quotes `"`   |
| Boolean        | `True` / `False`                        | `true` / `false` (lowercase) |
| None / Null    | `None`                                  | `null`                       |
| Trailing Comma | Allowed in latest Python versions       | Not allowed                  |

Example:

```python
# Python dict

# JSON equivalent
json_data = '{"active": true, "score": null}'
```

---

### **3. Usage**

* **Python dict** → Used **inside Python code** to store and manipulate data.
* **JSON** → Used to **send or store data** between programs, APIs, or servers.

* CSV files are **text files**, so they can be opened in **Excel, Notepad, or any text editor**.

```
Name,Age,Course

Maleesha,18,Java
Nimal,19,Math
```

* **Subsequent rows** contain the data.

---

### **CSV in Python**

Python has a built-in **`csv` module** to work with CSV files.

#### **Reading a CSV file:**

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  # each row is a list
```

#### **Writing to a CSV file:**

```python
import csv

data = [
    ["Name", "Age", "Course"],
    ["Shasidu", 17, "Python"],
    ["Maleesha", 18, "Java"]
]

with open("students.csv", "w", newline="") as file:

    writer.writerows(data)
```

### **Key Points**

4. In Python, `csv.reader` → read, `csv.writer` → write.
